```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python

def series_func(N: int):
    def fourier_saw(t):
        def term(n):
            costerm = np.cos(n * np.pi)
            sinterm = np.sin(n * t)
            return (costerm * sinterm) / n
        term = np.vectorize(term)
        series = term(np.linspace(1, N))
        summation = np.sum(series) * (-2 / np.pi)
        return summation
    return np.vectorize(fourier_saw)

def plot_and_save(func):
    T = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    f, ax = plt.subplots()
    ax.plot(func(T))
    f.savefig(f"images/fourier_{time.time_ns()}.png")
    plt.close(f)
```

```python
T = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
sawtooth = series_func(50)
plt.plot(sawtooth(T))

```

```python
N = np.linspace(1, 50)
# save all figures to the path with it's sequence number
[plot_and_save(series_func(n)) for n in N]
# close all plots to free system resources
plt.close('all')
```
